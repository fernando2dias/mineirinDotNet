```python
import re

T_KEYWORDS_INICIAL = "<keyword Trem>"
T_RESERVADOS = ["#intero", "#quebradin", "#charzin", "#trenzinDeChar", "//Prosa", "Oia"]
T_OPERADORES = "<op %s>"
T_OPERADORES_RELACIONAIS = "<op_relacional %s>"
T_OPERADORES_LOGICOS = "<op_logico %s>"
T_OPERADORES_RESERVADOS = ["TemBase", "TemGaioNao", "(", ")", "{", "}",".="] 
T_NUMERO_INTEIRO = "<numero %s>"
T_NUMERO_DECIMAL = "<numero %s>"
T_STRING = "<string %s>"
T_CHAR = "<char %s>"
T_IDENTIF = "<id %s>"

regex_variavel = re.compile(r"^[a-zA-Z].*$")
regex_numero = re.compile(r"^[0-9]+$")
regex_decimal = re.compile(r"^\d+,\d{1,8}?$")
regex_letra = re.compile(r"^'[a-zA-Z]'+$")
regex_string = re.compile(r'^"[a-zA-Z].*"')
regex_op = re.compile(r"!?=|[<>]=?")
regex_matematica = re.compile(r"(\d|[+/().-]|np\d+|\s+)*")


def afd_principal(token):
    if token == "Trem":
        return T_KEYWORDS_INICIAL
    elif token == "#intero": #INTEIRO
        return T_RESERVADOS[0]
    elif token == "#quebradin": #DECIMAL
        return T_RESERVADOS[1]
    elif token == "#charzin": #CHAR
        return T_RESERVADOS[2]
    elif token == "#trenzinDeChar": #STRING
        return T_RESERVADOS[3]
    elif token == "//Prosa": #COMENTARIO
        return T_RESERVADOS[4]
    elif token == "Oia": #PRINT
        return T_RESERVADOS[5]
    elif token == "+.+": #SOMA
        return T_OPERADORES % token
    elif token == "-.-": #SUBTRACAO
        return T_OPERADORES % token
    elif token == "/./": #DIVISAO
        return T_OPERADORES % token
    elif token == "*.*": #MULTIPLICACAO
        return T_OPERADORES % token
    elif token == "**.**": #POTENCIA
        return T_OPERADORES % token
    elif token == "%.%": #MODULO
        return T_OPERADORES % token
    elif token == "=.=": #IGUAL
        return T_OPERADORES_RELACIONAIS % token
    elif token == "\./": #DIFERENTE
        return T_OPERADORES_RELACIONAIS % token
    elif token == "<.<": #MENOR
        return T_OPERADORES_RELACIONAIS % token
    elif token == ">.>": #MAIOR
        return T_OPERADORES_RELACIONAIS % token
    elif token == ">.=": #MAIOR IGUAL
        return T_OPERADORES_RELACIONAIS % token
    elif token == "<.=": #MENOR IGUAL
        return T_OPERADORES_LOGICOS % token
    elif token == "&.&": #AND
        return T_OPERADORES_LOGICOS % token
    elif token == "^.^": #OR
        return T_OPERADORES_LOGICOS % token
    elif token == "!.!": #NOT
        return T_OPERADORES_LOGICOS % token
    elif token == "TemBase": #IF
        return T_OPERADORES_RESERVADOS[0]
    elif token == "TemGaioNao": #ELSE
        return T_OPERADORES_RESERVADOS[1]
    elif token == "(": #ABRE PARENTESES
        return T_OPERADORES_RESERVADOS[2]
    elif token == ")": #FECHA PARENTESES
        return T_OPERADORES_RESERVADOS[3]
    elif token == "{": #ABRE CHAVES
        return T_OPERADORES_RESERVADOS[4]
    elif token == "}": #FECHA CHAVES
        return T_OPERADORES_RESERVADOS[5]
    elif token == ".=": #ATRIBUIR = 
        return T_OPERADORES_RESERVADOS[6]
    elif re.fullmatch(regex_numero, token) : #INTEIRO
        return T_NUMERO_INTEIRO % token
    elif re.fullmatch(regex_decimal, token) : #DECIMAL
        return T_NUMERO_DECIMAL % token
    elif re.fullmatch(regex_string, token) : #STRING
        return T_STRING % token
    elif re.fullmatch(regex_letra, token) : #CHAR
        return T_CHAR % token
    elif re.fullmatch(regex_variavel, token) : #VARIAVEL
        return T_IDENTIF % token
    
    return None


arquivo = open('mineirin.txt', encoding='utf-8')
for l in arquivo.readlines():
    l = l.replace('\n','') # remove a quebra de linha
    if len(l) == 0:
        continue
    tokens = []
    for token in l.split():
        tokens.append(afd_principal(token))
    analise_lexica = open('análise_léxica.txt','a')
    analise_lexica.write(str(tokens)+'\n')

    print(tokens)
```
